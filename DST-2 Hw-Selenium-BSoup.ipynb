```python
data = []

names = browser.find_elements_by_xpath('/html/body/div[2]/div[4]/div[2]/div[2]/div/div[3]/div[2]/div')
# names

for i in range(1,len(names)):
    try:
        people = browser.find_element_by_xpath(f"/html/body/div[2]/div[4]/div[2]/div[2]/div/div[3]/div[2]/div[{i}]/div/div[2]/div/a").text
#     print(people)
    except:
        people = '-'
        
    try:
        pos = browser.find_element_by_xpath('/html/body/div[2]/div[4]/div[2]/div[2]/div/div[3]/div[2]/div[1]/div/div[2]/div/p/span[1]').text.split(':')
        position = pos[0]    
    except:
        position = '-'
        
    try:
        place = pos[1]
    except:
        place = '-'
        
    try: 
        phone = browser.find_element_by_xpath(f'/html/body/div[2]/div[4]/div[2]/div[2]/div/div[3]/div[2]/div[{i}]/div/div[1]/span').text
    except: 
        phone = '-'
        
    try: 
        email = browser.find_element_by_xpath(f'/html/body/div[2]/div[4]/div[2]/div[2]/div/div[3]/div[2]/div[{i}]/div/div[1]/a[1]').text
    except: 
        email = '-'
    
    try:
        link = browser.find_element_by_xpath(f'/html/body/div[2]/div[4]/div[2]/div[2]/div/div[3]/div[2]/div[{i}]/div/div[2]/div/a').get_attribute('href')
    except:
        link = '-'
    
    
    data.append([people, position, place, phone, email, link])

# print(data)

header = ['people', 'position', 'place', 'phone', 'email', 'link']
df = pd.DataFrame(data, columns = header)
df
```

```python
# Часть 2.

# Создаем функцию, принимающую на вход ссылку на аккаунт преподавателя, возвращающую строчку с началом года работы.
def person(url): 
    try: 
        browser.get(url)
        sleep(1.5)
        year = browser.find_element_by_xpath('/html/body/div[3]/div[3]/div[2]/div[3]/div/ul[2]/li[1]').text
    except: 
        year = '-'
    return year
```

```python
df['year'] = df.link.apply(person)
df
```
